```python
import google.generativeai as genai
import pytesseract
from PIL import Image

# Configure Gemini
genai.configure(api_key="")
model = genai.GenerativeModel("gemini-1.5-pro")

# Step 1: OCR extraction
image_path = "img1.png"
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
text_data = pytesseract.image_to_string(Image.open(image_path))

# Step 2: Combine OCR + image for Gemini reasoning
with open(image_path, "rb") as img:
    prompt = f"""
    The OCR text extracted from this diagram is:

    {text_data}

    Please explain what this diagram represents.
    Describe the process, relationships, and main ideas in a clear, structured format.
    """
    response = model.generate_content([img, prompt])

print(response.text)
```

```python
import json
import time
from serpapi import GoogleSearch

API_KEY = ""

# 10 diversified domains
domains = {
    "artificial_intelligence": "artificial intelligence",
    "renewable_energy": "renewable energy OR solar power OR wind turbine",
    "biotechnology": "biotechnology OR genetic engineering",
    "autonomous_vehicles": "autonomous vehicle OR self-driving car",
    "cybersecurity": "cybersecurity OR data encryption",
    "quantum_computing": "quantum computing OR quantum processor",
    "agriculture_technology": "precision agriculture OR smart farming",
    "medical_devices": "medical device OR wearable health monitor",
    "space_technology": "spacecraft OR satellite communication",
    "coffee_processing": "coffee processing OR food preservation"
}

for domain, query in domains.items():
    params = {
        "engine": "google_patents",
        "q": query,
        "api_key": API_KEY,
        "num": 50,   # up to 50 patents
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    
    filename = f"{domain}_patents.json"
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(results, f, ensure_ascii=False, indent=4)
    
    print(f"✅ Saved 50 patents for '{domain}' → {filename}")
    
    # Be polite to the API (avoid hitting rate limits)
    time.sleep(2)
```

```python
import json
import os
import requests

# Load the JSON file (replace 'patents.json' with your filename)
with open('autonomous_vehicles_patents.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Create a folder to store PDFs
pdf_folder = "autonomous_vehicles_patents"
os.makedirs(pdf_folder, exist_ok=True)

# Iterate through organic results and download PDFs
for result in data.get("organic_results", []):
    title = result.get("title", "untitled").replace(" ", "_").replace("/", "_")
    pdf_url = result.get("pdf")
    
    if pdf_url:
        local_filename = os.path.join(pdf_folder, f"{title}.pdf")
        try:
            print(f"Downloading {title}...")
            response = requests.get(pdf_url)
            response.raise_for_status()  # Raise error if download fails
            with open(local_filename, 'wb') as pdf_file:
                pdf_file.write(response.content)
            print(f"Saved to {local_filename}")
        except Exception as e:
            print(f"Failed to download {title}: {e}")

print("All downloads complete.")
```
